```python
import pandas as pd
from pymongo import MongoClient
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from collections import Counter

# MongoDB bağlantısı
client = MongoClient('mongodb://localhost:27017/')
db = client['climatewatch']

# Veri kaynakları
data_sources = {
    'climate_data': 'İklim Verisi (Sıcaklık, CO2 vb.)',
    'gdelt_events': 'GDELT Haber Eventleri',
    'disaster_events': 'Doğal Afetler',
    'policy_changes': 'İklim Politikası Değişiklikleri',
    'climate_summits': 'İklim Zirveleri',
    'google_trends': 'Google Trends',
    'open_datasets': 'Açık Veri Setleri',
    'kaggle_data': 'Kaggle Veri Setleri'
}

# Verileri çek
summary = {}
for col, label in data_sources.items():
    collection = db[col]
    cursor = collection.find({}, {'_id': 0})
    df = pd.DataFrame(list(cursor))
    summary[col] = df
    print(f'--- {label} ---')
    print(f'Toplam kayıt: {len(df)}')
    display(df.head())
```

```python
## 2. Her Veri Kaynağı İçin Trend ve Temel İstatistikler
for col, label in data_sources.items():
    df = summary[col]
    if len(df) == 0:
        continue
    
    # Tarih sütununu bul ve yıllık trendi çiz
    for date_col in ['date', 'year', 'Date', 'DATE']:
        if date_col in df.columns:
            years = pd.to_datetime(df[date_col], errors='coerce').dt.year
            year_counts = years.value_counts().sort_index()
            
            plt.figure(figsize=(10,6))
            sns.set_style('whitegrid')
            sns.lineplot(x=year_counts.index, y=year_counts.values, marker='o')
            plt.title(f'{label} - Yıllara Göre Kayıt Sayısı', fontsize=12, pad=15)
            plt.xlabel('Yıl', fontsize=10)
            plt.ylabel('Kayıt Sayısı', fontsize=10)
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()
            break
    
    # Temel istatistikler
    print(f'\n{label} - Temel İstatistikler:')
    display(df.describe(include='all'))
```

```python
## 3. GDELT Eventlerinde En Çok Geçen Kelimeler ve Kelime Bulutu
df = summary['gdelt_events']
if '27' in df.columns and '28' in df.columns and len(df) > 0:
    # Metinleri birleştir ve kelimeleri say
    all_text = df['27'].astype(str).str.cat(df['28'].astype(str), sep=' ').str.lower()
    words = ' '.join(all_text).split()
    most_common = Counter(words).most_common(30)
    
    # En çok geçen kelimeleri göster
    print('En çok geçen 30 kelime:')
    for word, count in most_common:
        print(f'{word}: {count}')
    
    # Kelime bulutu oluştur
    try:
        from wordcloud import WordCloud
        
        plt.figure(figsize=(15,8))
        wordcloud = WordCloud(width=1200, height=600, 
                            background_color='white',
                            max_words=100,
                            contour_width=3,
                            contour_color='steelblue').generate(' '.join(words))
        
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title('GDELT Eventlerinde En Çok Geçen Kelimeler', fontsize=16, pad=20)
        plt.tight_layout()
        plt.show()
    except ImportError:
        print('wordcloud kütüphanesi yüklü değil, kelime bulutu oluşturulamadı.')
```
