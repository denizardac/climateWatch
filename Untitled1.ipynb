```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
import pandas as pd

gdelt_df = pd.read_csv(
    r"C:/Users/ersan/OneDrive/Masaüstü/climateWatch-data-capture/gdelt_new/gdelt_climate_news_20150101_to_20201231.csv",
    sep=",",
    quotechar='"',
    engine="python",      # Python motoru
    encoding="latin-1"
)

print(gdelt_df.shape)
print(gdelt_df.columns.tolist())
```

```python
# DATE sütununu datetime'a çevir (YYYYMMDDHHMMSS formatı)
gdelt_df['DATE'] = pd.to_datetime(gdelt_df['DATE'], format='%Y%m%d%H%M%S')

# Yıl bilgisini year sütununa al
gdelt_df['year'] = gdelt_df['DATE'].dt.year

# Kontrol: İlk 5 satırı gör
print(gdelt_df[['DATE', 'year']].head())
```

```python
# Yıllık haber sayısını hesapla
news_count_per_year = (
    gdelt_df
    .groupby('year')
    .size()
    .reset_index(name='news_count')
)

# Kontrol için tabloyu göster
print(news_count_per_year)
```

```python
env_df = pd.read_csv(
    r"C:/Users/ersan/OneDrive/Masaüstü/climateWatch-data-capture/Enviro_data/merged_climate_data_1970_2020.csv"
)

# 2. CO2 ve deniz seviyesi sütunlarını lineer enterpolasyonla doldur
env_df[['co2_emissions', 'sea_level']] = (
    env_df[['co2_emissions', 'sea_level']]
    .interpolate(method='linear')
)

# 3. Kontrol: NaN kalmadı mı?
print(env_df[['year', 'temp_anomaly', 'co2_emissions', 'sea_level']].isna().sum())
```

```python
# 4. 2015–2020 arasını filtrele
env_sub = env_df[(env_df['year'] >= 2015) & (env_df['year'] <= 2020)].reset_index(drop=True)

# 5. Kontrol: Filtrelenmiş tablonun içeriği
print(env_sub)
```

```python
# news_count_per_year zaten hazır (2015-2020 aralığında)
# news_count_per_year = pd.DataFrame({'year': [...], 'news_count': [...]})

# 6. Birleştir
final_df = pd.merge(
    news_count_per_year,
    env_sub,
    on='year',
    how='inner'
)

# 7. Son tabloyu göster
print(final_df)
```

```python
# Korelasyon matrisini hesapla
corr = final_df[['news_count', 'temp_anomaly', 'co2_emissions', 'sea_level']].corr()

print(corr)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.scatter(final_df['news_count'], final_df['temp_anomaly'])
plt.xlabel('Haber Sayısı')
plt.ylabel('Sıcaklık Anomalisi (°C)')
plt.title('Haber Sayısı vs. Sıcaklık Anomalisi (2015–2020)')
plt.tight_layout()
plt.show()
```

```python
# Haber Sayısı vs CO2 Emisyonu
plt.figure(figsize=(6, 4))
plt.scatter(final_df['news_count'], final_df['co2_emissions'])
plt.xlabel('Haber Sayısı')
plt.ylabel('CO2 Emisyonu (kt)')
plt.title('Haber Sayısı vs CO₂ Emisyon (2015–2020)')
plt.tight_layout()
plt.show()

# Haber Sayısı vs Deniz Seviyesi
plt.figure(figsize=(6, 4))
plt.scatter(final_df['news_count'], final_df['sea_level'])
plt.xlabel('Haber Sayısı')
plt.ylabel('Deniz Seviyesi (cm)')
plt.title('Haber Sayısı vs Deniz Seviyesi (2015–2020)')
plt.tight_layout()
plt.show()
```

```python
corr_val = final_df['news_count'].corr(final_df['temp_anomaly'])
print(f"Haber sayısı ile sıcaklık anomalisi arasındaki Pearson korelasyonu: {corr_val:.3f}")
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Veri hazırlığı
X = final_df[['temp_anomaly']].values  # şekil (n,1)
y = final_df['news_count'].values      # şekil (n,)

# Model eğit
model = LinearRegression()
model.fit(X, y)
slope = model.coef_[0]
intercept = model.intercept_

print(f"Regresyon doğrusunun eğimi: {slope:.1f} (haber satır/°C)")
print(f"Regresyon sabiti: {intercept:.1f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.scatter(final_df['temp_anomaly'], final_df['news_count'], label='Veri')
# Doğrusu
x_line = np.linspace(final_df['temp_anomaly'].min(),
                     final_df['temp_anomaly'].max(), 100).reshape(-1,1)
y_line = model.predict(x_line)
plt.plot(x_line, y_line, color='red', label='Regresyon Doğrusu')
plt.xlabel('Sıcaklık Anomalisi (°C)')
plt.ylabel('Haber Sayısı')
plt.title('Baseline Regresyon: Haber Sayısı ~ Sıcaklık Anomalisi')
plt.legend()
plt.tight_layout()
plt.show()
```
